## II. Linear Algebra and  Probability Theory 

   ### A. Linear Algebra

Linear algebra is an important part of mathematics which being used in many fields of Science and engineering. Because of its continuous form , many people are still working on it. Before working on Machine Learning or deep learning we need to have better understanding of linear algebra. There are some mathematical objects in Linear Algebra.
 * Scalar - It is a number i.e. arrays of multiple numbers.
 * Vectors - It is an array of numbers that are arranged in orders and ususally identified by writing as x.
 * Matrices -  It is a 2D array of numbers i.e each element is recognised by two indices.
    for example,
 $
 x  = \begin{bmatrix}
  4.0 \\
  2.3 \\
  1.2 \\
  5.8
 \end{bmatrix}
 $
 
 * Tensors -  It is an array with more than axes which are arranged in a grid. It is denoted as __A__ .
 There are many other functions like transpose of matrix i.e. conversing the rows and columns with each other , adding matrices with same shape or multiplying scalar into a matrix (  _D_ - a .  _B_ + _c_ ).
 so when we multiply the two matrix if they have same shape i.e. equal number of rows and columns. 
 
 
 * Identity and Inverse Matrices
     There is a powerful tool in Linear algebra i.e Matrix inversion that allow us to analytically solve equations for many values of _A_ . The matrix inverse of _A_ is denoted as _A_$ ^-1 $ .

      The inverse matrix can only be represented with only little precision on a digital computer , algorithms.
 * Linear Dependence and Span 
      A set of vector is linear dependent if one vector is linear combination of other vector else they are linear independent.
      The span of a set of vectors is the set of all points obtainable by liner combination of the original vectors. 
 * Norms 
     Norms are the function used to measure the size of vectors in machine learning.
     
     Squared norm is convenient to work with mathematically and computationally the normal norm itself. There are many types of norm like max norm and frobenius norm used in machine learning.
     
There are many other types of matrices ( Diagonal Matrices - consists of zero and non zero entries only along the main diagonal , symmetric matrix - matrix that is equal to its own transpose ( _A_ - _A_$ ^\top $ ), orthogonal matrix - it is a square matrix where rows and columns are mutually orthonormal).

* Eigendecomposition
    Eigenvectors are the set of vectors connected with a set of linear system i.e characteristics vectors or latent vectors.So the decomposition of a matrix into matrices consisting of eigenvectors and eigenvalues is called as eigendecomposition of that vector.
    
    
    the eigendecomposition is unique onlyif all the eigenvalues are unique.

* Singular Value Decomposition 
    This is another way of decomposition of a matrix into singular vectors and values. Every matrix has a singular value decomposition, but the same is not true of the eigenvalue decomposition. The most useful feature of the SVD is that we can use it to partially generalize matrix inversion to nonsquare matrices. 

The Moore- Penrose pseudoinverse is matrix inverse that is a unique _n_ x _m_ matrix pesudoinverse.

We can derive PCA - principal components analysis using all these knowledge of basic linear algebra.

Linear algebra is one of the fundamental mathematical disciplines necessary tounderstanding deep learning.



   ### B. Probability Theory 

Probability theory is a mathematical framework for solving some uncertain conditions. In the Machine Learning and AI, we use this concept for how to use these technology and to analysie the behavior of proposed AI system.

Probability theory is a important tool for many disciplines in the field of science and engineering.

Machine learning must always deal with uncertain quantities and sometimes stochastic quantities. 

There are three possible sources of uncertainty: 1.Inherent stochasticity in the system being modeled, 2.Incomplete observability, 3.Incomplete modeling.

we also sometime use probability to represent a degree of belief. 

* Probability Distribution 
    Aprobability distributionis a description of how likely a random variable orset of random variables is to take on each of its possible states. The way we describe probability distributions depends on whether the variables are discrete or continuous.
    
* Discrete Variables and Probability Mass Functions
    A probability distribution over discrete variables may be described using aproba-bility mass function(PMF).
    
* Continuous Variables and Probability Density Functions
    We use continuous random variables that describe probability distributions using a probability density function (PDF) rather than a probability mass function.
    
* Marginal Probability
    The probability distribution over the subset is known as the marginal probability distribution.
    
* Conditional Probability
    The probability of some event given that some other event has happened is called a conditional probability.

The variance gives a measure of how much the values of a function of a random variable _x_ vary as we sample diï¬€erent values of _x_ from its probability distribution.

The square root of the variance is known as the standard deviation.

Several simple probability distributions are useful in many contexts in machine learning.

* Bernoulli Distribution
* Multinoulli Distribution
* Gaussian Distribution
* Exponential and Laplace Distributions
* The Dirac Distribution and Empirical Distribution


* Information Theory
    Information theory is a branch of applied mathematics that revolves around finding out how much information is present in a signal.The basic intuition behind information theory is that learning that an unlikely event has occurred is more informative than learning that a likely event has occurred.
    
    If we have two separate probability distributions _P(x)_ and _Q(x)_ over the same random variable _x_, we can measure how diï¬€erent these two distributions are using the __Kullback-Leibler (KL) divergence__.
    
Therefore the above basic concepts of probability theory and linear algebra that are most relevant to deep learning and machine learning.

### Classification

__So for Classification I chose " Google, Amazon and more Employee Reviews" dataset from UCI Machine Learning Repository (https://archive.ics.uci.edu/ml/datasets/bank+marketing). The data is related with direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or not ('no') subscribed.__

There are 17 variables and in total 45211 records . The folowing are the attribute Information:

Input variables:

1 - age (numeric) 

2 - job : type of job (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')

3 - marital : marital status (categorical: 'divorced','married','single','unknown'; note: 'divorced' means divorced or widowed)

4 - education (categorical: 'basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course','university.degree','unknown')

5 - default: has credit in default? (categorical: 'no','yes','unknown')

6 - housing: has housing loan? (categorical: 'no','yes','unknown')

7 - loan: has personal loan? (categorical: 'no','yes','unknown')

8 - contact: contact communication type (categorical: 'cellular','telephone') 

9 - month: last contact month of year (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')

10 - day: last contact day.

11 - duration: last contact duration, in seconds (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no').

12 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)

13 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)

14 - previous: number of contacts performed before this campaign and for this client (numeric)

15 - poutcome: outcome of the previous marketing campaign (categorical: 'failure','nonexistent','success')

16 - y : has the client subscribed a term deposit? (binary: 'yes','no')

17 - Balance : balance in the bank account (numeric)
